```python
import numpy as np
import cv2
import os
import sys
import matplotlib.pyplot as plt
```

```python
print(cv2.__version__)
```

```python
 #image file read
path = "images/"
#path = '/content/drive/My Drive/ComputerVision/images/'
fname= "cat.png"
filepath = path+fname
#이미지 파일 읽기
image = cv2.imread(filepath)

if image is None:
    print('Image load failed!')
    sys.exit()
    
#파일의 크기 확인
imgHeight, imgWidth, channel = image.shape
print(fname, imgHeight, imgWidth, channel)

# image를 Widnows로 출력
cv2.imshow(fname, image)

# 칼라 이미지를 그레이스케일로 변환
# Using cv2.COLOR_BGR2GRAY color space 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow(fname+"- gray", gray_image)

# ESC 키 입력 시 Windows 닫음
cv2.waitKey(0)
cv2.destroyAllWindows() 

#파일로 저장
fname_g = filepath+"_gray.png"
cv2.imwrite(fname_g, gray_image)
```

```python
#matplot 으로 출력
import matplotlib.pyplot as plt

#image file read
#path = '/content/drive/My Drive/ComputerVision/images/'
#(1)‘lenna.bmp’파일을 칼라로 읽기
path="images/"
fname= "lenna.bmp"
filepath = path+fname

#이미지 파일 읽기
imageBGR = cv2.imread(filepath)

print('type(image):', type(image))
print('image.shape:', image.shape)

if len(image.shape) == 2:
       print('image is a grayscale image')
elif len(image.shape) == 3:
        print('image is a truecolor image')

#(2)BGR 을 RGB로 변환하여 matplotlib의 imshow()로 출력
imageRGB = cv2.cvtColor(imageBGR, cv2.COLOR_BGR2RGB)

plt.imshow(imageBGR)
plt.title("BGR color image")
plt.show()

plt.imshow(imageRGB)
plt.title("RGB color image")
plt.show()

#(3)‘lenna.bmp’파일을 그레이 스케일로 읽고 출력
#gray image read
gray_image = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)

plt.imshow(gray_image)
plt.title("gray scale image")
plt.show()
plt.imshow(gray_image, cmap = "gray")
plt.title("gray scale image")
plt.show()


#(4)‘lenna.bmp’파일을 그레이 스케일로 변환하여 ‘lenna_gray.bmp’ 로 저장
#image write(save)
gray_image = cv2.cvtColor(imageRGB, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap = "gray")
plt.title("gray scale image")
plt.show()
fname = path + "lenna_g.bmp"
cv2.imwrite(fname,gray_image)
```
